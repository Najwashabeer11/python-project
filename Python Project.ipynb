```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_df= pd.read_excel("myexcel.xlsx")
data_df.head()
```

```python
data_df.describe(include='all')
```

```python
data_df.info()
```

```python
#Generate random numbers between 150 and 180 by replacing incorrect data in height column


data_df['Height']=np.random.randint(150,180,size=len(data_df))
data_df
```

```python
data_df
```

```python
#How many employees belong to each team and calculate the percentage of employees in each team relative to the total number of employees
#No.of employees
tot_num_employees=data_df['Name'].count()
print(tot_num_employees)

#num.of employees in each team
num_in_each_team=data_df['Team'].value_counts()
print(num_in_each_team)

#percentage splitting
percentage_splitting=(num_in_each_team/tot_num_employees)*100
print(percentage_splitting)

#pie chart representation
plt.figure(figsize=(8,8))
plt.pie(num_in_each_team.values,labels=num_in_each_team.index,autopct='%1.1f%%')
plt.title('Percentage splitting wrt employees')
plt.show()
```

```python
#segregate the emplyees w.r.t to different positions
position_distribution=data_df['Position'].value_counts()
print(position_distribution)

#Bar chart representation
plt.bar(position_distribution.index,position_distribution.values,width=0.5,color='r',edgecolor='black')
plt.title('Number of employees in each position')
plt.xlabel('Position')
plt.ylabel('Number of employees')
plt.figure(figsize=(6,3))
plt.show()
```

```python
#identifying age group
age_data=data_df['Age'].value_counts()
age_data1=pd.DataFrame({'Age':age_data.index,'Number of employees':age_data.values})
print(age_data1)
max=age_data1['Number of employees'].max()

#Histogram representation
plt.hist(data_df['Age'],bins=[10,20,30,40,50],color='y',edgecolor='black')
plt.title('Age group analysis')
plt.xlabel('Age group')
plt.ylabel('Number of employees')
plt.show()
```

Max number of employees between age group of 20 and 30

```python
#identifying which team and position spending high salary
team_salary=data_df.groupby(['Team','Position'])['Salary'].sum().idxmax()
print(team_salary)
```

```python
#Age salary correlation and represent it visually
correlation=data_df[['Age','Salary']].corr()
print(correlation)
#Scatterplot chart
sns.scatterplot(x='Age',y='Salary',data=data_df)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age and salary correlation')
plt.show()
```
